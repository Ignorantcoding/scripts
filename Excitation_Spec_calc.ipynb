```python
xx = np.linspace(0,(math.pi/2),70)
both_x=[]
mzm_x=[]
mpm_x=[]
none_x=[]
both_y=[]
mzm_y=[]
mpm_y=[]
none_y=[]
store_q=[]
error= 1e-4
for J in range(0,70):
    J_2 = xx[J]
    #x = np.linspace(0,(math.pi/2),70)
    eigen_2=[]
    for h_2 in range(0,70):
        z=0 #zero mode
        p=0 #pi mode
        U_T= np.dot(lin.expm(2*xx[h_2]*H_1), lin.expm(2*J_2*H_2))
        eigen= list(np.sort(np.linalg.eigvals(U_T)))
        temp=[] #Stores all values of phase for a combo of J_2 and h_2
        for eig in eigen:
            q=((cmath.log(eig)/(-1j))/math.pi)
            temp.append(q)
            if q.real <= error and q.real >= -error: #If phase is approx 0
                z=1
            elif q.real <= 1+error and q.real >= 1-error: #If phase approx 1
                p=1
            elif q.real <= -1+error and q.real >= -1-error: #If phase approx 1
                p=1
                
        store_q.append([temp,xx[h_2],J_2])      
        if p==1:
            if z==1:
                both_x.append(xx[h_2]/math.pi)
                both_y.append(J_2/math.pi)
            else: #i.e p=1 and z=0, only pi mode
                mpm_x.append(xx[h_2]/math.pi)
                mpm_y.append(J_2/math.pi)
        elif z==1:
            mzm_x.append(xx[h_2]/math.pi)
            mzm_y.append(J_2/math.pi)
        else:
            none_x.append(xx[h_2]/math.pi)
            none_y.append(J_2/math.pi)
```

```python
#Taking delta h from 0 to 0.175 pi
#accepted error to save quasivalue
h= 0.375*np.pi
J_2= 0.15*np.pi
error= 0.5
for delta_h in np.linspace(0,0.175*np.pi,70):
    #Repeat for each delta h 80 times, and take average of all these
    #We'll save each entry around +-1 in these 2 lists
    quasi_pos=[]
    quasi_neg=[]
    for i in range(0,80):
    #Now construct the matrix H1 by selecting constant h_i randomly from interval [h+delta h, he-delta h]
        above_dig, below_dig =[],[]
        for i in range(0,49):
            if i%2==0:
                #select random numbers from interval
                above_dig.append(2*random.uniform(h+delta_h, h-delta_h))
                below_dig.append(-2*random.uniform(h+delta_h, h-delta_h))
            else:
                above_dig.append(0)
                below_dig.append(0)
        H_1_dis = (tridiag(below_dig, above_dig))
        #After getting H1, we now compute quasi spectrum
        #get U_T
        U_T= np.dot(lin.expm(H_1_dis), lin.expm(2*J_2*H_2))
        #Get eigenvalues of U_T
        eigenvalues= list(np.sort(np.linalg.eigvals(U_T)))
        #Get quasi-spectrum values and save only those near 1 and -1 by taking max and min quasivalues as all quasiavlues are either at 1/-1 or below
        quasi=[]
        for eig in eigenvalues:
            q= ((cmath.log(eig)/(-1j))/math.pi)
            quasi.append(q.real)
        #q1 is closest to being -1
        q1= min(quasi)
        #q2 is closest to being 1
        q2= max(quasi)
        if q2 <= 1+error and q2 >= 1-error: #If phase is approx 0
            quasi_pos.append(q2)
        if q1 <= -1+error and q1 >= -1-error: #If phase is approx 0
            quasi_neg.append(q1)
    #Now take average of values
    if len(quasi_pos)>0 and len(quasi_neg)>0:
        quasi_1= sum(quasi_pos)/len(quasi_pos)
        quasi_2= sum(quasi_neg)/len(quasi_neg)
        plt.scatter([delta_h]*2, [quasi_1,quasi_2], c= 'r', marker= '.')
    else:
        if len(quasi_pos)>0:
            quasi_1= sum(quasi_pos)/len(quasi_pos)
            plt.scatter(delta_h, quasi_1, c= 'r', marker= '.')
        elif len(quasi_neg)>0:
            quasi_2= sum(quasi_neg)/len(quasi_neg)
            plt.scatter(delta_h, quasi_2, c= 'r', marker= '.')
plt.xlabel('Delta_h')
plt.ylabel('Quasienergy Excitation Spec')
```
